```python
# Doc:
# https://redis-limpyd.readthedocs.io/en/latest/index.html
# Code:
# https://github.com/limpyd/redis-limpyd
# Redis as a service:
# https://app.redislabs.com/#/login
from limpyd import model
```

```python
main_database = model.RedisDatabase(
    host="redis-16634.c250.eu-central-1-1.ec2.cloud.redislabs.com",
    password="XXX",
    port=16634,
    db=0
)
```

```python
class Bike(model.RedisModel):
    database = main_database
    # name = model.InstanceHashField(indexable=True, unique=True)
    name = model.StringField(indexable=True, unique=True)
    color = model.InstanceHashField()
    wheels = model.StringField(default=2)
```

```python
mountainbike = Bike(name="mountainbike")
mountainbike.wheels.get()
```

```python
mountainbike.wheels.incr()
mountainbike.wheels.get()
```

```python
mountainbike.name.set("tricycle")
# tricycle = Bike.collection(name="tricycle")[0]
tricycle = Bike.collection(name="tricycle").instances()[0]
tricycle
```

```python
tricycle.wheels.get()
```

```python
tricycle.hmset(color="blue")
```

```python
tricycle.hmget('color')
```

```python
tricycle.hmget('color', 'name')
```

```python
tricycle.color.hget()
```

```python
tricycle.color.hset('yellow')
```

```python
tricycle.hmget('color')
```

```python
mountainbikes = Bike.collection(name="mountainbike")
mountainbikes_instances = Bike.collection(name="mountainbike").instances()
if len(mountainbikes_instances) > 0:
  mountainbike = mountainbikes_instances[0]
  mountainbike.delete()
```

```python
tricycles = Bike.collection(name="tricycle")
tricycles_instances = Bike.collection(name="tricycle").instances()
if len(tricycles_instances) > 0:
  tricycle = tricycles_instances[0]
  tricycle.delete()
```
